### Information on Simple ITK and numpy indexing

http://insightsoftwareconsortium.github.io/SimpleITK-Notebooks/Python_html/03_Image_Details.html 

Conversion between numpy and SimpleITK
- SimpleITK and numpy indexing access is in opposite order!
- SimpleITK: image[x,y,z]
- numpy: image_numpy_array[z,y,x]












Read the scan into the code

```python
if path.endswith('.nii') or path.endswith('.nii.gz') or path.endswith('nrrd'):
    image       = sitk.ReadImage(path)
    array       = np.swapaxes(sitk.GetArrayFromImage(image), 0, 2)
    ConstPixelSpacing = image.GetSpacing()
    bone        = core.thresholdnp(array, 1100, 1500)
    softtissue  = core.thresholdnp(array, 0, 80)
    
elif path.endswith('.dcm'):
    pathHead    = os.path.split(path)[0]
    reader      = sitk.ImageSeriesReader()
    dicomNames = reader.GetGDCMSeriesFileNames(pathHead)
    reader.SetFileNames(dicomNames)
    image       = reader.Execute()
    array       = np.swapaxes(sitk.GetArrayFromImage(image), 0, 2)
    ConstPixelSpacing = image.GetSpacing()
    bone        = core.thresholdnp(array, 900, 1500)
    softtissue  = core.thresholdnp(array, 0, 80)
```

Reshape the scan so that circles can be found easily

```python
# Reshaping 
downsample_fraction = 1  # change to downsample
arrayReshaped = core.reshape(array, np.divide(ConstPixelSpacing, downsample_fraction))  
boneReshaped = core.reshape(bone, np.divide(ConstPixelSpacing, downsample_fraction))
softtissueReshaped = core.reshape(softtissue, np.divide(ConstPixelSpacing, downsample_fraction))
```

Swap round what defines eye number 1 and eye number 2 if necessary

```python
cc = 0.5*(c1Reshaped + c2Reshaped)
xcoord = cc[1]
ycoord = cc[0]
c1, c2 = np.divide(c1Reshaped, ConstPixelSpacing), np.divide(c2Reshaped, ConstPixelSpacing)

# adjusting so that c1 is always less than c2 in the x coordinate, if the eyes are in the upper half of the scan
# adjusting so that c1 is always greater than c2 in the x coordinate, if the eyes are in the lower half of the scan

# upper half 
if xcoord>=boneReshaped.shape[0]/2:
    if c1[1]<c2[1]: # x coords of c1
        pass 
    else: 
        temp = c2Reshaped
        c2Reshaped = c1Reshaped
        c1Reshaped = temp
# lower half 
else:
    if c1[1]>c2[1]: # x coords of c1
        temp = c2Reshaped
        c2Reshaped = c1Reshaped
        c1Reshaped = temp 
    else: 
        pass
ccReshaped = 0.5*(c1Reshaped+c2Reshaped)

angle1, angle2 = findeyes.anglesFromEyes(c1Reshaped, c2Reshaped, boneReshaped.shape)
```

Rotate the scan according to the skew of the scan; rotate so that the eyes now lie in the same slice

```python
from scipy.ndimage.interpolation import rotate

while angle1<=-45: #skews are never that big
    angle1 = angle1+90
    
while angle1>=45: #skews are never that big
    angle1 = angle1-90
    
rotatedSofttissue = rotate(softtissueReshaped, -angle1, mode='nearest', axes=(0,2)) # want angle1 to be 90
rotatedBone = rotate(boneReshaped, -angle1, mode='nearest', axes=(0,2)) # want angle1 to be 90
```

v2 is a unit vector going from one eye to the other (in the rotated scan)

```python
v2 = c1ReshapedRotated-c2ReshapedRotated
v2 = np.divide(v2, np.linalg.norm(v2))
```

now use an optimiser to find the relationship between the scan that has had its skew corrected for, and the original scan. Save this rotation + translation in an array T

```python
import scipy
def f(params, args):
    
    a,b,c,d,e,f,h,i,j = params
    vector =  args[0][0]
    comparison = args[1][0]

    T = np.asarray([[a, b, c],
                    [d, e, f],
                    [h, i, j]])
    
    v1 = np.dot(T, vector)
    
    v2 = comparison
    diff = np.linalg.norm(v1-v2)
    return diff

crr = (c1ReshapedRotated-c2ReshapedRotated)

if np.linalg.norm(c1ReshapedRotated-c1Reshaped)>np.linalg.norm(c1ReshapedRotated-c2Reshaped):
    temp = c1Reshaped
    c1Reshaped = c2Reshaped
    c2Reshaped = temp
else:
    pass

cr = (c1Reshaped-c2Reshaped)

c = np.divide(cr, ConstPixelSpacing)

x = scipy.optimize.minimize(f,([1,0,0,0,1,0,0,0,1]), args=([[crr], [c]]))

a,b,c,d,e,f,h,i,j = x.x

T = np.asarray([[a, b, c],
                [d, e, f],
                [h, i, j]])
```

Find the actual normal of the midplane of the original scan; this means that the T matrix must be multiplied with the "reshapedNormal" (the normal found for the midplane of the skew-corrected scan) to get it into the original coordinate system. The normal is then made a unit vector

```python
normal = np.dot(T, reshapedNormal)
normal = np.divide(normal, np.linalg.norm(normal))
```

```python
a,b,c = normal
coord = 0.5*(c1+c2)
d = np.dot(normal, coord)
a,b,c,d
```

```python
def saveMask(array, path, name, image):
    mask = array.astype(np.float64)
    mask = np.swapaxes(mask, 0,2)
    maskimage = sitk.GetImageFromArray(mask)
    maskimage.SetDirection(image.GetDirection())
    maskimage.SetOrigin(image.GetOrigin())
    maskimage.SetSpacing(image.GetSpacing())
    savePath = os.path.join(os.path.split(path)[0], '{}.nii.gz'.format(name))
    sitk.WriteImage(maskimage,savePath)
```
